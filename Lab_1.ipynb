#**Tokenization of words and sentences & Text Normalization**

```python
import nltk
```

## **Tokenization of words**

```python
from nltk.tokenize import word_tokenize
nltk.download('punkt_tab')

text = "God is Great! I won a lottery."
print(word_tokenize(text))
```

#### For printing third word only:



```python
words = word_tokenize(text)
print(words[2])
```

##**Tokenization of sentences**

```python
from nltk.tokenize import sent_tokenize
print(sent_tokenize(text))
```

####For printing second sentence only:

```python
s = sent_tokenize(text)
print(s[1])
```

## **Normalizing text**

#### Converting all characters to lowercase:

```python
[w.lower() for w in gberg_sent_tokens]
```

#### Removing Stop Words & Punctuation:


```python
stpwrds = stopwords.words('english') + list(string.punctuation)
print(stpwrds)
```

#### Removing Stop Words & Punctuation with a list comprehension:

```python
[w.lower() for w in gberg_sent_tokens if w.lower not in stpwrds]

# Correct ones:
# [w.lower() for w in word_tokenize(gberg_sent_tokens[0]) if w.lower() not in stpwrds]

# gberg_tokens = word_tokenize(gutenberg.raw())
# [w.lower() for w in gberg_tokens if w.lower() not in stpwrds]
```

##**Porter Stemmer**

```python
raw = """ DENNIS: Listen, strange women lying in ponds distributing swords
    is no basis for a system for government. Supreme executive power derives from
    a mandate from the masses, not from some farcial aquatic ceremony."""

tokens = word_tokenize(raw)
porter = nltk.PorterStemmer()
[porter.stem(t) for t in tokens]
```

##**Lanster Stemmer**

```python
lancaster = nltk.LancasterStemmer()
[lancaster.stem(t) for t in tokens]
```

## **Snowball Stemmer**

```python
snowball = nltk.SnowballStemmer("english")
[snowball.stem(t) for t in tokens]
```

## **Lemmatization**

```python
nltk.download('wordnet')
from nltk.tokenize import word_tokenize
tokens = word_tokenize(raw)
wnl = nltk.WordNetLemmatizer()
[wnl.lemmatize(t) for t in tokens]
```
